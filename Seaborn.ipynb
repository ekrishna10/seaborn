```python
#How to working joint plots
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
j = sns.jointplot(data = movies, x ='CriticRating' , y ='AudienceRating',kind='scatter')
plt.show()
```
